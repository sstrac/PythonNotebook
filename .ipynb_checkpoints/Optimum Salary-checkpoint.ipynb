## Calculating Optimum Salary after 50K

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

def calculate_income_tax(gross_salary):
    if gross_salary > 12500 and gross_salary < 50001:
        return gross_salary * 0.2
    elif gross_salary >= 50001:
        return (gross_salary - 50000) * 0.4 + 50000 * 0.2
    else: return 0

def calculate_national_insurance(gross_salary):
    return gross_salary * 0.138 if gross_salary > 732*12 else 0

def negate_costs(gross_salary):
    income_tax_cost = calculate_income_tax(gross_salary)
    national_insurance_cost = calculate_national_insurance(gross_salary)
    student_debt = 300    
    return out if (out := gross_salary - income_tax_cost - national_insurance_cost - student_debt) > 0 else 0        

x = list(range(0, 80000, 1000))
y = list(map(negate_costs, x))

plt.plot(x,y)
plt.xlabel('Gross Income')
plt.ylabel('After Costs')
plt.show()
```
